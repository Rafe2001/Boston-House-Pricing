```python
plt.figure(figsize=(12,5))
plt.title('Countries with highest CO2 emissions (metric tons per capita)')
plt.ylabel('CO2 emissions (metric tons per capita) in 2019')
df.groupby('Country Name')['2019'].sum().sort_values(ascending=False)[0:15].plot(kind='line', marker='*', color='red', ms=10)
df.groupby('Country Name')['2019'].sum().sort_values(ascending=False)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.title('Countries with highest CO2 emissions (metric tons per capita)')
plt.ylabel('CO2 emissions (metric tons per capita) in 2018')
df.groupby('Country Name')['2018'].sum().sort_values(ascending=False)[0:15].plot(kind='line', marker='*', color='red', ms=10)
df.groupby('Country Name')['2018'].sum().sort_values(ascending=False)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

```python
df.rename(columns={"Country Name":"country_name"}, inplace=True)
```

```python
plt.figure(figsize=(10,30))
sns.barplot(x='2019',y='country_name',data=df, order=df.sort_values('2019',ascending=False).country_name)
plt.title("Co2 emission per capita by country")
```

```python
df_transpose = df.transpose()      
```

```python
df_test = pd.concat([df_transpose.iloc[0:1, :], df_transpose.iloc[4:35, :]])
df_test.reset_index(inplace=True)
df_test.columns = df_test.iloc[0]
df_test.rename(columns={"country_name":"year"}, inplace=True)
df_test.drop([0], inplace=True)
year = df_test['year']
df_test.drop(['year'], axis=1, inplace = True)
for i in df_test.columns:
    df_test[i] = pd.to_numeric(df_test[i])
df_test = df_test.divide(df_test.sum(axis=1), axis=0)
df_test = pd.concat([year, df_test], axis = 1)
df_test
```

```python
df_selected = df_test[['Qatar',
                          'Bahrain',
                          'United Arab Emirates',
                          'Brunei Darussalam',
                          'Canada',
                          'Saudi Arabia',
                          'United States',
                          'year']]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

for i in df_selected.columns:
    if(i == "year"):
        i = 0
    else:    
        ax = sns.lineplot(data=df_selected, x='year', y=df_selected[i], label=i)
        
        ax = sns.scatterplot(data=df_selected, x='year', y=df_selected[i])

plt.ylabel("CO2 emissions (metric tons per capita)")
plt.title("CO2 emissions by countries")
```

```python
df_selected.set_index('year', inplace=True)
df_selected.index.name = None
df_selected
```

```python
sns.set_theme(palette="inferno")
plt.rcParams["figure.figsize"] = (10, 10)
df_selected.plot.area()
plt.legend(loc='upper right')

plt.ylabel(" % total of CO2 emissions (metric tons per capita)")
plt.title("CO2 emissions per capta by country")

plt.show
```
